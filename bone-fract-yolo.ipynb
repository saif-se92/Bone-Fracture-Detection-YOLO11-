```python
from ultralytics import SAM
import matplotlib.pyplot as plt
from IPython.display import Image, display
import os # For path manipulations

# Load your YOLO model
yolo_model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

# Define the image path
image_path = "/kaggle/input/bone-fracture-detection-computer-vision-project/BoneFractureYolo8/test/images/image1_117_png.rf.e200e6b84f70bd27ba1e1123d052755e.jpg"

# Run batched inference on the image (results is a list of YOLO Results objects)
yolo_results_list = yolo_model(image_path)

# Load your SAM model (ensure "sam2_b.pt" is accessible in your Kaggle environment)
# If "sam2_b.pt" is not a standard model name that Ultralytics downloads automatically,
# ensure the file path is correct (e.g., /kaggle/input/your-sam-model-dataset/sam2_b.pt)
sam_model = SAM("sam2_b.pt")

# Process each result from YOLO
for yolo_result in yolo_results_list:
    class_ids = yolo_result.boxes.cls.int().tolist()
    
    # Check if YOLO detected any objects
    if class_ids: 
        boxes = yolo_result.boxes.xyxy # Get bounding boxes from YOLO

        # Run SAM model on the original image with YOLO's bounding boxes
        # The 'save=True' argument will save the output image to a 'runs/segment/predict*' directory.
        # sam_output_list is a list of SAM Results objects (usually one for a single input image).
        sam_output_list = sam_model(yolo_result.orig_img, bboxes=boxes, verbose=False, save=True, device=0)

        if sam_output_list:
            sam_result = sam_output_list[0] # Get the first (and likely only) SAM result

            # --- Option 1: Display using the .plot() method (Recommended for notebooks) ---
            # The .plot() method returns a NumPy array of the image with segmentations.
            print("Displaying SAM output using .plot():")
            segmented_image_array = sam_result.plot()
            
            plt.figure(figsize=(10, 10))
            plt.imshow(segmented_image_array)
            plt.axis('off')
            plt.title("SAM Segmented Image")
            plt.show()

            # --- Option 2: Display the saved image file ---
            # The image is saved because of 'save=True'.
            # sam_result.save_dir contains the directory where the image was saved.
            # The filename is usually the same as the input image's filename.
            print(f"\nSAM output was saved in directory: {sam_result.save_dir}")
            
            original_filename = os.path.basename(yolo_result.path) # Get filename from original YOLO input path
            saved_image_path = os.path.join(sam_result.save_dir, original_filename)
            
            print(f"Attempting to display saved image from: {saved_image_path}")
            if os.path.exists(saved_image_path):
                display(Image(filename=saved_image_path))
            else:
                print(f"Error: Saved image not found at {saved_image_path}.")
                print("Files in save directory:")
                for f_name in os.listdir(sam_result.save_dir):
                    print(f"- {f_name}")
                    # If your image has a generic name like 'image0.jpg', you might need to find it
                    # if f_name == "image0.jpg": # Or a more robust way to identify it
                    #    display(Image(filename=os.path.join(sam_result.save_dir, f_name)))

        else:
            print("SAM model did not produce any output for this image.")
    else:
        print(f"YOLO model did not detect any objects in the image: {yolo_result.path}")
```
